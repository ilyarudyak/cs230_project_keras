## elastic deformation

### train data - no crop

```python
original_size, target_size = 512, 512
```

```python
data = ISBI2012()
```

```python
train_gen = data.generator(batch_size=1,  mode='training')
```

```python
n = 3
fig, ax = plt.subplots(n, 4, figsize=(15, 15))
[axi.set_xticks([]) for axi in ax.ravel()]
[axi.set_yticks([]) for axi in ax.ravel()]
for i in range(n):
    image_batch, mask_batch = next(train_gen)
    ax[i, 0].imshow(data.image_data[i].reshape(original_size, original_size), cmap='gray')
    ax[i, 1].imshow(image_batch.reshape(target_size, target_size), cmap='gray')
    ax[i, 2].imshow(data.mask_data[i].reshape(original_size, original_size), cmap='gray')
    ax[i, 3].imshow(mask_batch.reshape(target_size, target_size), cmap='gray')
```
